```python
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

```

```python
load_dotenv()  # loads variables from .env
project=os.getenv('GOOGLE_CLOUD_PROJECT')  # reads the variable

#model
llm=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    vertexai=True,
    project=project

)
```

```python
#define basic tools
from langchain_core.tools import tool

```

```python
# create and register the tools
@tool
def add(a:int|float , b:int|float) -> int:
    """
    add two number
     args:
     a(int|float):first argument
      b(int|float):second argument

      return:
       int|float:a+b
    """
    return a+b
```

```python
#make llm to aware of this tools
llm_with_tools=llm.bind_tools([add])
```

```python
question="what is the result of 4+5"
```

```python
response_without_tools=llm.invoke(question)
```

```python
#response
response_without_tools.pretty_print()
```

```python
response_with_tools=llm_with_tools.invoke(question)
```

```python
#response
response_with_tools.pretty_print()
```

```python
#now add more tools

@tool
def subtract(a:int|float , b:int|float) -> int|float:
    """
    subtract second number from first number
     args:
     a(int|float):first argument
      b(int|float):second argument

      return:
       int|float:a-b
    """
    return a-b

@tool
def multiply(a:int|float , b:int|float) -> int|float:
    """ multiply two number
     args:
     a(int|float):first argument
      b(int|float):second argument

      return:
       int|float:a*b
    """
    return a*b

@tool
def divide(a: int | float, b: int | float) -> int | float:
    """
    divide two numbers

    args:
        a (int|float): first argument
        b (int|float): second argument

    return:
        int|float: a/b

    raise:
        ValueError: if b is 0
    """
    if b == 0:
        raise ValueError("divide by zero is not allowed")

    return a / b

@tool
def modulus(a:int|float ,b:int|float)->int|float:
    """ find remainder when first number is devided by second number

    Args:
    a(int|float):first argument
    b(int|float):second argument

    Return :
    int|float:a%b

    Raise:
    ValueError: if b is 0

    """

    if b==0:
        raise ValueError("modulus by zero is not allowed")
    return a%b
        
```

```python
#lets bind multiple tool to llm
llm_with_tools = llm.bind_tools([add, subtract, multiply, divide, modulus])
```

```python
question="""i have taken 10000 rupees from a friend for  monthly interest rate of 1.5% 
i want find out the amount which i will end up paying 
if i can return the amount int he 10 month , it is simple interest"""
```

```python
response_with_tools=llm_with_tools.invoke(question)
```

```python
response_with_tools.pretty_print()
```
